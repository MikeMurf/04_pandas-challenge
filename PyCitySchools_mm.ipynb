```python
#    import dependencies
import pandas as pd

#    load files to analyse
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

#    read school and student data files into pandas dataframes
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#    merge the pandas dataframe a single dataframe 
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#    print the first five records of the dataframe
school_data_complete.head()
```

```python
#    calculate the total number of schools
unique_school_name=school_data_complete['school_name'].unique()

total_schools_number=len(unique_school_name)

total_schools_number
```

```python

#    calculate the total number of students
total_students_number=school_data_complete["student_name"].count()
total_students_number
```

```python
#    calculate the total budget of all schools
total_budget = school_data["budget"].sum()
total_budget
```

```python
#   calculate the percentage of students with a passing maths score (70 or greater)

#   find the  students with a passing maths score (70 or greater)
students_passing_maths = school_data_complete.loc[school_data_complete["math_score"] >= 70]

#    count the total number of all students who sat for maths
number_students_passing_maths = students_passing_maths["Student ID"].count()

#   calculate the percentage of students with a passing maths score (70 or greater)
percent_passing_maths = (number_students_passing_maths / total_students_number) * 100

#    save the percentage of students with a passing maths score (70 or greater)
percent_passing_maths
```

```python
#   calculate the percentage of students with a passing reading score (70 or greater)

#   find the  students with a passing reading score (70 or greater)
students_passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]

#    count the total number of all students who sat for reading
number_students_passing_reading = students_passing_reading["Student ID"].count()

#   calculate the percentage of students with a passing reading score (70 or greater)
percent_passing_reading = (number_students_passing_reading / total_students_number) * 100

#    save the percentage of students with a passing maths score (70 or greater)
percent_passing_reading
```

```python
#   calculate the percentage of students with a passing score (70 or greater) in both maths and reading

#   find the  students with a passing score (70 or greater) in both maths and reading
number_students_passing_both = school_data_complete[(school_data_complete["math_score"] >= 70)
                        & (school_data_complete["reading_score"] >= 70)]

#    count the total number of all students
number_students_passing_both = number_students_passing_both["Student ID"].count()

#   calculate the percentage of students with a passing maths score (70 or greater)
percent_passing_both = (number_students_passing_both / total_students_number) * 100

#    save the percentage of students with a passing maths score (70 or greater)
percent_passing_both
```

```python
#    create a district summary dataframe to hold the above results
district_summary = pd.DataFrame({
    "Total Schools": f"{total_schools_number:}",
    "Total Students": f"{total_students_number:}",
    "Total Budget": f"${total_budget:,.2f}",
    "Average Math Score": f"{average_maths_score:.6f}",
    "Average Reading Score": f"{average_reading_score:.5f}",
    "% Passing Math": f"{percent_passing_maths:.6f}",
    "% Passing Reading": f"{percent_passing_reading:.6f}",
    "% Overall Passing": f"{percent_passing_both: .6f}"
}, index=[0])

district_summary
```

```python
#    create a set of all schools by school type
school_type = school_data.set_index('school_name')['type']
school_type
```

```python
#    calculate total students per school
total_students_per_school = school_data_complete["school_name"].value_counts()
total_students_per_school
```

```python
#    calculate total budget per school
total_school_budget = school_data.set_index('school_name')['budget']
total_school_budget
```

```python
#    calculate the per student budget
per_student_budget = (school_data.set_index('school_name')['budget']/school_data.set_index('school_name')['size'])
per_student_budget
```

```python
#    calculate the average maths score per school
average_school_maths_score = school_data_complete.groupby(["school_name"]).mean()["math_score"]
average_school_maths_score
```

```python
#    calculate the average reading score per school
#average_school_reading_score = school_name['reading_score'].mean()
average_school_reading_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
average_school_reading_score
```

```python
#    calculate the percentage of students who passed maths for each school
school_maths_percent_pass = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/total_students_per_school*100
school_maths_percent_pass
```

```python
#    calculate the percentage of students who passed reading for each school
school_reading_percent_pass = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/total_students_per_school*100
school_reading_percent_pass
```

```python
#   find the  students with a passing score (70 or greater) in both maths and reading for each school

percent_passing_both = school_data_complete[(school_data_complete["reading_score"] >= 70)
                                               & (school_data_complete["math_score"] >= 70)]

percent_passing_both = percent_passing_both.groupby(["school_name"]).count()["student_name"] / total_students_per_school * 100

percent_passing_both
```

```python
#    create a school dataframe to hold the above results
school_summary = pd.DataFrame({"School Type": school_type,
                                   "Total Students": total_students_per_school,
                                   "Total School Budget": total_school_budget,
                                   "Per Student Budget": per_student_budget,
                                   "Average Math Score": average_school_maths_score,
                                   "Average Reading Score": average_school_reading_score,
                                   "% Passing Math": school_maths_percent_pass,
                                   "% Passing Reading": school_reading_percent_pass,
                                   "% Overall Passing": percent_passing_both})


#     data wrangling
school_summary = school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                                         "Average Math Score", "Average Reading Score", 
                                         "% Passing Math", "% Passing Reading", 
                                         "% Overall Passing"]]

school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)

#     display the dataFrame
school_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

```python
#    create a table that lists the average maths score for students of 9th, 10th, 11th and 12th grade level at each school

ninth_grade_maths = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
tenth_grade_maths = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
eleventh_grade_maths = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
twelfth_grade_maths = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()


grade_maths_scores = pd.DataFrame({
        "9th": ninth_grade_maths,
        "10th": tenth_grade_maths,
        "11th": eleventh_grade_maths,
        "12th": twelfth_grade_maths
})

grade_maths_scores = grade_maths_scores[['9th', '10th', '11th', '12th']]

grade_maths_scores.index.name = "School Name"

#    format the average maths scores table
grade_maths_scores.style.format({'Average Math Score': '{:.2f}', 
                              'Average Reading Score': '{:.2f}', 
                              '% Passing Math': '{:.2f}', 
                              '% Passing Reading':'{:.2f}', 
                              '% Overall Passing': '{:.2f}'})

grade_maths_scores.head(5)

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#    The requirement is to "create a table that breaks down school performances based on average spending ranges (per student),
#    and to use 4 reasonable bins to group school spending".

#    the bins used are <$585, $585-$615, $615-$645 and $645-$675 spending per student per annum
bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#    set up spending bins data frame
school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = group_names)

#    group by spending
by_spending = school_data_complete.groupby('spending_bins')

#    calculate the average maths score for the school
average_maths_score = by_spending['math_score'].mean()

#    calculate the average reading score for the school
average_reading_score = by_spending['reading_score'].mean()

#    calculate the average spending per student for those students who passed maths
average_spend_per_maths_pass = school_data_complete[school_data_complete['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100

#    calculate the average spending per student for those students who passed reading
average_spend_per_read_pass = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100

#    calculate the average spending per student for those students who passed both maths and reading
average_spend_per_both_pass = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100
    
#   set up the average spending per student dat frame          
scores_by_spend = pd.DataFrame({
    "Average Math Score": average_maths_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": average_spend_per_maths_pass,
    "% Passing Reading": average_spend_per_read_pass,
    "% Overall Passing": average_spend_per_both_pass            
})

scores_by_spend.index.name = "Per Student Budget"

#    format the average spending per student table
scores_by_spend.style.format({'Average Math Score': '{:.2f}', 
                              'Average Reading Score': '{:.2f}', 
                              '% Passing Math': '{:.2f}', 
                              '% Passing Reading':'{:.2f}', 
                              '% Overall Passing': '{:.2f}'})
```
